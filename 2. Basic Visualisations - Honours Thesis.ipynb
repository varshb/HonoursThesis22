## Percentile Graphs: Annual

```python
#percentiles of interest
percentiles = np.array([0.95, 0.9 , 0.85, 0.8 , 0.75, 0.5 , 0.25, 0.2 , 0.15, 0.1 , 0.05])
colours = np.array(['red','orange', 'green', 'brown', 'blue', 'black','blue', 'brown', 'green','orange','red'])

#create a separate dataset that includes only years from 1968 to 2021 inclusive
annualData = df
```

```python
#Sydney
#line graph of the annual mean max temperature
ax = (((annualData[(annualData.location == 'Sydney')].
   groupby("year").mean()["meanMaxTemp"].
   plot(marker = 'o', color = 'black', figsize = (14,8))
  )))

#horizontal percentile lines
for p,colour in zip(percentiles, colours):
    #calculate percentile for annual figures
    quantile = annualData[(annualData.location == 'Sydney')].groupby("year").mean()["meanMaxTemp"].quantile(p)
    #plot the dashed lines
    ax.axhline(y=quantile, color=colour, linestyle='--', label = p)

ax.set_xticks(np.arange(1965,2022,5))
ax.legend(bbox_to_anchor=(1.0, 1))
ax.set_ylabel('Temperature (C)')
ax.set_xlabel('Year')


plt.title("Sydney Annual")    
plt.show()
```

## Empirical Distributions

```python
#https://stackoverflow.com/questions/3209362/how-to-plot-empirical-cdf-ecdf

def ecdf_values(x):    
    # Sort values and find length
    x = np.sort(x)
    n = len(x)
    # Create percentiles
    y = np.arange(1, n + 1, 1) / n
    return x, y


def ecdf_plot(x1, x2, labels, name = 'Value', plot_normal = False):
    """
    ECDF plot of x

    Params
    --------
        x (array or list of numerics): distribution for ECDF
        name (str): name of the distribution, used for labeling
        plot_normal (bool): plot the normal distribution (from mean and std of data)
    
    Returns
    --------
        none, displays plot
    
    """
    xsList = []
    ysList = []
    xs1, ys1 = ecdf_values(x1)
    xs2, ys2 = ecdf_values(x2)
    xsList.append(xs1)
    xsList.append(xs2)
    ysList.append(ys1)
    ysList.append(ys2)
    colours = [colourP[i] for i in [0,3]]

    # create the figure and axes
    fig, ax = plt.subplots(1, 1, figsize=(14, 5))

    # unpack all the axes subplots
    
    for i, colour in zip(np.arange(2), colours):
        ax.scatter(xsList[i], ysList[i], c= colour, label = labels[i]);

        plot_range = ax.get_xlim()[1] - ax.get_xlim()[0]
        fig_sizex = fig.get_size_inches()[0]
        data_inch = plot_range / fig_sizex
        right = 0.6 * data_inch + max(xsList[i])
        gap = right - max(xsList[i])
        left = min(xsList[i]) - gap

        if plot_normal:
            gxs, gys = ecdf_values(np.random.normal(loc = xsList[i].mean(), 
                                                    scale = xsList[i].std(), 
                                                    size = 100000))
            ax.plot(gxs, gys, 'orange');

        ax.vlines(x=min(xsList[i]), 
                   ymin=0, 
                   ymax=min(ysList[i]), 
                   color = 'b', 
                   linewidth = 2.5)

        # Add ticks
        plt.xticks(size = 16)
        plt.yticks(size = 16)
        ax.set_xlabel("Temperature (C)")
        ax.set_ylabel("Percentile") 
        ax.legend(loc = 'upper left')
        
        # annotations
        ps = [0.25, 0.5, 0.75]

        for p in ps:

            ax.set_xlim(left = left, right = right)
            ax.set_ylim(bottom = 0)
            value1 = x1.quantile(p)
            ax.hlines(y=p, xmin=left, xmax = value1, linestyles = ':', colors = 'black', linewidth = 2);
            ax.vlines(x=value1, ymin=0, ymax = p, linestyles = ':', color = colours[0], linewidth = 2)
            ax.text(x = p / 3, y = p - 0.01,transform = ax.transAxes, s = f'{int(100*p)}%', size = 15, color = 'black', alpha = 0.7)
            ax.text(x = value1 + 0.05, y = 0.01, size = 15, horizontalalignment = 'left', s = f'{value1:.2f}', color = colours[0], alpha = 0.8);
            
            value2 = x2.quantile(p)
            ax.hlines(y=p, xmin=value1, xmax = value2, linestyles = ':', colors =  'black', linewidth = 2);
            ax.vlines(x=value2, ymin=0.1, ymax = p, linestyles = ':', color =  colours[1], linewidth = 2)
            ax.text(x = value2 + 0.05, y = 0.09, size = 15, horizontalalignment = 'left', s = f'{value2:.2f}', color =  colours[1], alpha = 0.8);

        # fit the labels into the figure
        ax.set_title(name, size = 20)
```
